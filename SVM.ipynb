```python
import numpy as np
import pandas as pd
from os import listdir
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
arousal_labels = np.load('save_arousal_labels.npy')
valence_labels = np.load('save_valence_labels.npy')
data = np.load('save_data.npy')
arousal_labels=arousal_labels.reshape(32*40)
valence_labels=valence_labels.reshape(32*40)
data=data.reshape([32*40,40*101])
```

```python
def split(data,labels,rate=0.8):
#     data = np.concatenate(data,axis=0)
#     labels = np.concatenate(labels,axis=0)
#     print(data.shape)
#     print(labels.shape)
    np.random.seed(999)
    index = np.arange(labels.shape[0])
    index_random = index
    
    np.random.shuffle(index_random)
    data = data[index_random]
    labels = labels[index_random]
    
    test_data = data[int(data.shape[0]*rate):]
    test_labels = labels[int(data.shape[0]*rate):]
    
    train_data = data[:int(data.shape[0]*rate)]
    train_labels = labels[:int(data.shape[0]*rate)]
#     print(train_data.shape)
    return train_data,train_labels,test_data,test_labels
```

寻找最佳超参

```python
def GridSearch(data,labels):
    parameters={'kernel':('rbf','sigmoid'),'C':[1,5,10,20,50,100]}
    svr=SVC()
    clf=GridSearchCV(svr,parameters)
    clf.fit(data,labels)
    print(clf.best_estimator_)
```

```python
GridSearch(data,arousal_labels)
```

```python
def ClassTest(data,labels,c=1,kernel='rbf'):
    traind,trainl,testd,testl=split(data,labels)
    clf = SVC(c,kernel)
    clf.fit(traind,trainl)
    errorCount = 0.0
    classifierResult = clf.predict(testd)
    for i in range(len(testl)):
        realResult = testl[i]
        print("分类返回结果为%d\t真实结果为%d" % (classifierResult[i],realResult))
        if(classifierResult[i] != realResult):
            errorCount += 1.0
    print("总共错了%d个数据\n错误率为%f%%" % (errorCount, errorCount/len(testl) * 100))
```

```python
ClassTest(data,arousal_labels,c=50,kernel='rbf')
```

```python
GridSearch(data,valence_labels)
```
